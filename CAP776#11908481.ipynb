```python
set1={1,2,3,4,5}
set2={1,2,3,6,8}
print(set1.symmetric_difference(set2)) #toprint unique values among two 2 sets i.e. symmetric difference
```

```python
#Update the set by adding elements from an iterable/another set.
set1={1,2,3,4,5}
set2={1,2,3,6,8,11,2,56,87}
set1.update(set2)  # to add elements of one set to another or perform UNION_ON_SET
print(set1,set1|set2) # | can also be used  
```

```python
#Update the set by keeping only the elements found in it and an iterable/another set.
set1={1,2,3,4,5}
set2={1,2,3,6,8}
set1.intersection_update(set2)  # to perform INTERSECTION_ON_SET
print(set1,set1&set2)           # & is also used for same purpose
```

```python
#Update the set by removing elements found in an iterable/another set.
set1={1,2,3,4,5}
set2={1,2,3,6,8}
set1.difference_update(set2)  # to remove elements from set 1 which r found in set 2
print(set1,set1-set2)           # - is also used for same purpose
```

```python
#Update the set by only keeping the elements found in either set, but not in both.
set1={1,2,3,4,5}
set2={1,2,3,6,8}
set1.symmetric_difference_update(set2)  # to get only those elements which are found only in 1 set
print(set1) # or (set1^set2)           # ^ is also used for same purpose
```

# CLASS AND OBJECT

```python
class Student(object):
    def __init__(self,name,roll_no): # class constructor
        self.name=name;              # initialization
        self.roll_no=roll_no;
    def input(self,nam,roll):        # method to take input
        self.name=nam;
        self.roll_no=roll
    def output(self):                # method to print data values
        print("Roll: ",self.roll_no,"\tName: ",self.name) 
```

```python
st=Student("nipun",47) # object of the class
st.output()
st.input("xyz",15)     # calling the class method
st.output()
```

```python
class circle():
    def __init__(self,radius): # class constructor
        self.radius=radius;              # initialization
    def getArea(self):        # method to take input
        area=3.14*self.radius*self.radius
        return area        
    def getCircumference(self):                # method to print data values
        print("circumference: ",2*3.14*self.radius) 
sr=circle(3)
print(sr.getArea())
sr.getCircumference()
```

```python
class Temp():
    #def __init__(self,fh,cel): # class constructor
     #   self.fh=fh
      #  self.cel=cel
    def convertCelcius(self,fh):        # method to take input
        tocel = ((fh-32)*5)/9
        return tocel        
    def convertFahrenhiet(self,cel):                # method to print data values
        tofah =((9/5*(cel))+32)
        return tofah
sr=Temp()#98.6,37)
print(sr.convertCelcius(98.7),sr.convertFahrenhiet(37))
```

# Direction clockwise

```python
def turn_clockwise(direction):
    if direction in('e','E'):
        print ("S")
    elif direction in('s','S'):
        print ("W")
    elif direction in('w','W'):
        print ("N")
    elif direction in('n','N'):
        print ("E")
    else:
        print("None")
dir=input("enter direction ")
turn_clockwise(dir)      
    
```
